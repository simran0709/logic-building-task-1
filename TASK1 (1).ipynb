```python
#2. Pass / Fail Analyzer
marks=[45, 78, 90, 33, 60]
pass_count=0
fail_count=0
for mark in marks:
  if mark >= 50:
     pass_count +=1
  else:
     fail_count +=1
print("total number of pass students are:",pass_count)
print("total number of fail students are:",fail_count)
```

```python
#3. Simple Data Cleaner
names = [" Alice ", "bob", " CHARLIE "]
cleaned_names=[]
for name in names:
  cleaned=name.strip().lower()
  cleaned_names.append(cleaned)
print(cleaned_names)
```

```python
#4. Message Length Analyzer
messages = ["Hi", "Welcome to the platform", "OK"]
for message in messages:
    length=len(message)
    print(f"{message}-->length:{length}")
    if length > 10:
       print("long message")
```

```python
#5. Error Message Detector
logs = ["INFO", "ERROR", "WARNING", "ERROR"]
error=0
for log in logs:
  if "ERROR" in log:
      error +=1
print("total number of errors are:",error)
```
